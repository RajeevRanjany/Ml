```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("/content/Placement_Dataset.csv")
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

Central tendencies:
1. Mean
2. Median
3. Mode


```python
#Analyse the distribution of data in the salary column
fig,ax = plt.subplots(figsize=(8,8))
sns.histplot(dataset['salary'], kde=True)
plt.show()
```

Replacing Missing Values with median values

```python
dataset['salary'].fillna(dataset['salary'].median(), inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.head()
```

Dropping the missing values : When we have large dataset then we can drop some records

```python
salary_dataset = pd.read_csv("/content/Placement_Dataset.csv")
salary_dataset.head()
```

```python
salary_dataset.isnull().sum()
```

```python
#drop missing values
salary_dataset = salary_dataset.dropna(how='any')
```

```python
salary_dataset.isnull().sum()
```
