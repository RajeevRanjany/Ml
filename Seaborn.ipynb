Toy Dataset seaborn has some build-in dataset

```python
# Load the 'tips' dataset from seaborn
tips = sns.load_dataset("tips")
display(tips.head())
```

```python
sns.regplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'time', hue='smoker', style ='smoker', size = 'size')
```

```python
# house price dataset
from sklearn.datasets import fetch_california_housing

house_california = fetch_california_housing(as_frame=True)

house = pd.DataFrame(house_california.data, columns=house_california.feature_names)
house['PRICE'] = house_california.target
```

```python
house.head()
```

```python
 print(house)
```

```python
#Distribution Plot
sns.displot(house['PRICE'])
```

```python
sns.distplot(house['PRICE'])
```

Importannt : Correlation Matrix

```python
correlation = house.corr()
```

```python
#Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar='true',square = 'true',fmt='.1f', annot = True, annot_kws={'size':8},cmap='Blues')
```
