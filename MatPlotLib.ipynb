Matplot Library : Useful for making Plots


```python
import matplotlib.pyplot as plt
```

```python
#for random value importing this
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the Data

```python
#sin wave
plt.plot(x,y)
plt.show
```

```python
#cosine wave
plt.plot(x,z)
plt.show
```

```python
plt.plot(x,y,z)
plt.show()
```

```python
#adding title x axis and y title
plt.plot(x,y)
plt.title('sin wave')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#parabola
x = np.linspace(-10,10,100)
y = x**2;
plt.title('Parabola :- ')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'gx')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

Bar Graph

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['Japanese','French','EngLish','Latin','Bhojpuri']
people = [23,17,35,29,12]
plt.xlabel('Languages')
plt.ylabel('No of Speaking People')
ax.bar(langs,people)
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
langs = ['Japanese','French','EngLish','Latin','Bhojpuri']
people = [23,17,35,29,12]
ax.pie(people, labels=langs, autopct='%1.2f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,20)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')

# Data
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)

# 3D scatter plot
ax.scatter(x, y, z, c=z, cmap='Blues')

# Axis labels
ax.set_xlabel('X axis (sin)')
ax.set_ylabel('Y axis (cos)')
ax.set_zlabel('Z axis (random values)')
ax.set_title('3D Scatter Plot')

plt.show()
```
